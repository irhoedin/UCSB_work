# Carrier density and Seeback Coefficient

$$ S = -\frac{k_{B}}{e}\ln\frac{2(1-\rho)}{\rho} $$

where $\rho = N/N_{0} $. $N$ is the number of electrons and $N_{0}$is the number of the sites.

by Mori p.146 or DOI: 10.1103/physrevb.13.647 .

What is the shape of the Sâ€“n plot?

```python
%matplotlib inline
from import_all import *

KB = 1.38e-23 # Boltzman constant [J/K]
Q_elem = 1.6e-19 # Elementary charge [C]

def Seeback(n, n0):
    
    rho = (n/n0)
    S = -(KB/Q_elem) * np.log((2 * (1 - rho))/rho)
    
    return S

N0 = 1e20
N = np.logspace(12,20)

S = Seeback(N, N0)

f, ax = plt.subplots()
ax.plot(N, S, 'o')
ax.set_xscale('log')
ax.set_xlabel('Carrier density [cm-3]')
ax.set_ylabel('S [V/K]')
```
