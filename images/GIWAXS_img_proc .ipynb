```python
%matplotlib inline
from PIL import Image
import numpy as np
from matplotlib import pylab as plt

img = np.array( Image.open('GIWAXS_image.tif') , 'f')
plt.imshow( img )
```

```python
y_num, x_num = img.shape # get pixcel number
y = np.linspace(1,0,y_num + 1) #add '1' to the pixcel number to have a compatible meshgrid 
x = np.linspace(0,1,x_num + 1)
X, Y = np.meshgrid(x, y)
```

```python
f, ax = plt.subplots()
mesh = ax.pcolormesh(X, Y, img)
plt.colorbar(mesh)
mesh.set_clim(vmin=2.5, vmax=5) #designating z-value
```
